```julia
# plot complex numbers
```

```julia
using PyPlot
```

```julia
real_n = -2:0.01:2;
imag_n = 2:-0.01:-2;
```

```julia
# initialize array for complex numbers
fz = zeros(ComplexF64,length(real_n),length(imag_n));
```

```julia
# f(z) = (a + bi)^2 + 1, where z = a + bi
for m = 1:length(real_n)
    for n = 1:length(imag_n)
        fz[m,n] = (real_n[m] + imag_n[n]*im).^2 +1;
    end
end
```

```julia
# plot a graph of f(z)
surf(real_n,imag_n,abs.(fz),facecolors=get_cmap("viridis")(angle.(fz)))
xlabel("real number")
ylabel("imaginary number")
zlabel("magnitude")
title("complex plane")
```
